```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
import re
import jieba
import langconv
```

```python
fname = 'D:\python\datasource\Lession_05\sqlResult_1558435.csv'
```

```python
news = pd.read_csv(fname, encoding = 'gb18030')
len(news)
```

```python
news.head()
```

```python
news['cut_content'] = ''
```

#### 切词

```python
content_id = []
content_source = []
content_source_ = []
content_text = []
for i, content in enumerate(news['content']):    
    id, source, news_content = news['id'][i], news['source'][i], news['content'][i]
    text = str(news_content)
    text = text.strip()
    if not text: continue
    text = langconv.Converter('zh-hans').convert(text)
    text = ''.join(re.findall(r'[^\\n]', text))
    text = ''.join(re.findall(r'[\d|\w]+',text))
    text = ' '.join(jieba.cut(text))
    content_id.append(id)    
    content_text.append(text)
    content_source.append(source)
    if i % 2000 ==0:
        print('Saved ' + str(i) + ' lines')    
```

#### 保存切词后的文件

```python
cutFile = 'D:\python\datasource\Lession_05\cutResult.csv'
```

#### 保存切词后的文件

```python
data = pd.DataFrame({'id':content_id, 'source':content_source, 'content':content_text})
data.to_csv(cutFile,encoding = 'gb18030')
```

#### 加载切词后的文件

```python
cut_content = pd.read_csv(cutFile, encoding = 'gb18030')
cut_content = cut_content.loc[:, ~cut_content.columns.str.contains('^Unnamed')]

for c in cut_content:
    print(c)
```

```python
cut_content = cut_content.fillna(' ')
```

```python
len(cut_content)
```

```python
cut_content.head()
```

#### 新华社和非新华社索引

```python
xinhua_index = []
unxinhua_index = []
y = []
for i, source in enumerate(cut_content['content']):
    if '新华社'  in str(cut_content['source'][i]):
        xinhua_index.append(i)
        y.append(1)
    else:
        unxinhua_index.append(i)
        y.append(0)
```

```python
len(xinhua_index),len(unxinhua_index)
```

#### 训练和测试数据索引

```python
xinhua_data = np.random.choice(xinhua_index,size = 6000, replace = False).tolist()
unxinhua_data = np.random.choice(unxinhua_index,size = 6000, replace = False).tolist()
train_index = xinhua_data[:3000] + unxinhua_data[:3000]
random.shuffle(train_index)
test_index = xinhua_data[3000:] + unxinhua_data[3000:]
random.shuffle(test_index)
```

```python
len(train_index), len(test_index)
```

#### 文本向量化

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
stop_words = {}
```

```python
train_text = [cut_content['content'][i] for i in train_index] 
test_text = [cut_content['content'][i] for i in test_index] 
```

```python
text = train_text
tfidf = TfidfVectorizer(max_features=20000, stop_words=stop_words)
fit_tfidf = tfidf.fit_transform(text)
train_x = fit_tfidf.toarray()
test_x = tfidf.transform(test_text).toarray()
```

```python
print(tfidf.get_feature_names(), end = ' ')
```

```python
# 训练集
train_y = [y[i] for i in train_index]
```

```python
# 测试集
test_y = [y[i] for i in test_index]
```

#### 结果评估

```python
from sklearn import metrics 
```

```python
def measures(pred):    
    print("accuracy      : %0.3f" % metrics.accuracy_score(test_y, pred))
    print("precision score: %0.3f" % metrics.precision_score(test_y, pred))
    print("recall score  : %0.3f" % metrics.recall_score(test_y, pred))
    print("f1 score      : %0.3f" % metrics.f1_score(test_y, pred))
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr_pred = lr.fit(train_x,train_y).predict(test_x)
measures(lr_pred)
```

### SVM

```python
from sklearn.svm import LinearSVC
```

```python
svm = LinearSVC()
svm_pred = svm.fit(train_x, train_y).predict(test_x)
measures(svm_pred)
```

### Bayes

```python
from sklearn.naive_bayes import GaussianNB, BernoulliNB, MultinomialNB , ComplementNB
```

```python
gnb = GaussianNB()
gnb_pred = gnb.fit(train_x, train_y).predict(test_x)
measures(gnb_pred)
```

```python
bnb = BernoulliNB()
bnb_pred = bnb.fit(train_x, train_y).predict(test_x)
measures(bnb_pred)
```

```python
cnb = ComplementNB()
cnb_pred = cnb.fit(train_x, train_y).predict(test_x)
measures(cnb_pred)
```

```python
mnb = MultinomialNB()
mnb_pred = mnb.fit(train_x, train_y).predict(test_x)
measures(mnb_pred)
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
rfc_pred = rfc.fit(train_x, train_y).predict(test_x)
measures(rfc_pred)
```

### XGBoost

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
abc = AdaBoostClassifier()
abc_pred = abc.fit(train_x, train_y).predict(test_x)
measures(abc_pred)
```

### KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn_pred = knn.fit(train_x, train_y).predict(test_x)
measures(knn_pred)
```
